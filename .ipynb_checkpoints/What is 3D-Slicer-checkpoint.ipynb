# What is 3D Slicer?
___

3D slicer 는 :
+ 의료영상들을 분석(registration and interactive segmentation)과 시각화(volume randering)하고, image-guided therapy(IGT) 영역에서의 리서치를 위한 소프트웨어 플렛폼이다.
+ 다중의 os에서 사용가능한 무료, 오픈 소스 소프트웨어이다. (Linux, MacOSX 그리고 windows)
+ 확장 가능하고, 알고리즘과 애플리케이션들을 추가하는 강력한 플러그인 기능들을 제공한다.

아래의 특징들을 포함한다 :
+ 다중 기관 : 머리부터 발 끝까지
+ MRI, CT, US, 핵의학, 현미경 등을 포함하는 다기능 영상을 지원
+ 장치들의 양방향 인터페이스 지원

사용에 어떠한 제한도 없지만, Slicer는 임상용으로는 승인되지 않고 연구용으로 제작되었다. 권한 및 해당 규칙 준수는 사용자의 책임이다. license에 대한 더 자세한 내용은 여기를 참조 하라 (https://www.slicer.org/wiki/License)

# Hardware Requirements
___
3D slicer는 Mac, Linux 그리고 windows에서 사용가능한 오픈 소스 패키지이다. 3D slicer를 실행시키기 위해서는 컴퓨터가 원래 이미지 데이터와 프로세스 결과를 저장 할 수 있는 그래픽 기능과 메모리를 가지고 있어야 한다. 또한 64bit 시스템이 필요하다. 자세한 내용은 여기를 참조하라. 

(https://www.slicer.org/wiki/Documentation/4.10/SlicerApplication/HardwareConfiguration)

# Main Application GUI
___
3D Slicer는 모듈형 아키텍쳐를 기반으로 구축되어있다. 주 애플리케이션 GUI는 'menu bar', 'Application Toolbar', 'Module GUI Panel', 'Data Probe Panel', '2D Slice viewers', '3D viewer' 등의 6개의 구성요소로 나뉜다. 이 section은 주 애플리케이션 GUI의 기본 기능들에 대해 소개할 것이다. 자세한 정보는 링크를 참조하라 
(https://www.slicer.org/wiki/Documentation/4.10/SlicerApplication/MainApplicationGUI)

3D Slicer를 열고 당신의 데이터 혹은 검색을 통한 샘플 데이터를 로드 한다. 계속하여 user interface를 클릭하라

![1.png](./png_files/1.png)

welcome panel에서, 당신의 데이터 혹은 검색을 통한 샘플 데이터를 로드할 수 있다. 당신의 데이터가 없다면, 샘플 데이터는 3D Slicer의 특징들을 탐색할 수 있어 유용하다.

![2.png](./png_files/2.png)

자세한 옵션을 보려면 각 슬라이스 뷰어 혹은 3D 뷰어의 왼쪽 상단 모서리에 있는 푸쉬 핀을 클릭하라. Slice viewers에서 수평 막대를 사용하여 slice를  스크롤 하거나 선택할 수 있다. 로드된 데이터 혹은 다운받은 샘플 데이터를 사용하여 다양한 옵션을 탐색 할 수 있다.

# Modules
___

3D slicer는 다양한 기능을 할 수 있는 모듈들이 풍부하다. 포괄적인 모듈 목록은 documentation page를 참조하라.
(https://www.slicer.org/wiki/Documentation/4.10#Modules_by_category ) 

모듈들은 각각의 모듈에 대한 정보가 있고, tutorial들을 포함하고 있는 documentation page를 가지고 있다. 3D slicer는 모듈 드롭다운 메뉴의 상단 부분에 표시되는 10개 이상의 core module들이 있다.

## Core modules
___
#### Welcome 
: 3D slicer가 시작되면 켜지는 기본 모듈이다. 패널에는 데이터를 로드하고 3D Slicer의 사용자 정의를 하는 옵션이 있다. 이 옵션들을 아래에는 3D slicer를 사용하기 위한 필수 정보가 들어있는 drop-down box들이 있다.

#### Annotations
: 주석 또는 보충 정보의 생성 및 편집 허용. 현재, rulers(자)와 roi(region of interest)가 지원된다. 

#### Markups
: scene과 관련된 markup들을 생성하고 수정하는 것을 허용한다. 현재, fiducial의 list들은 markup 형태로 지원된다.

#### Data
: 현재의 scene안의 개체들을 나열하고, 검색, 이름변경, 삭제 그리고 이동 과 같은 기본 기능들을 수행할 수 있다.

#### DataStore
: 사용자가 데이터 세트를 다운로드하고 업로드할 수 있도록 허용한다.

#### DICOM
: CTK 및 DCMT에서 지원하는 DICOM을 통합한다.

#### Editor
: volume을 수동으로 분할 할 수 있다.

#### Models
: 모델들의 display 파라미더들을 로드하고 조정한다. 사용자가 3D표면 모델들의 외관을 변경하고 구성할 수 있다.

#### Scene Views
: 장면 내 데이터의 다중 "실시간 views"를 구성하기 위한 도구이다. 사용자는 view들을 여러개 생성할 수 있고, 파라미터들을 조정할 수 있다.

#### Subject Hierarchy
: 이 모듈은 slicer에서 central data-organizing point 역할을 한다. 주체 계층 노드는 cloning(복제), bulk transforming(대량 변환), bulk show/hide(대량 표시/숨김), type-specific features(유형별 기능들) 및 delete(삭제)와 rename(이름 변경) 과 같은 기본 노드 기능들을 제공한다. 

#### Transforms
: 이 모듈을 변환 매트릭스를 만들고 편집하는데 사용된다. 변환 가능한 목록에서 변환된 목록으로 노드를 이동하거나, 데이터 모듈의 변환 노드 아래에서 노드를 끌어 이러한 관계를 설정할 수 있다.

#### Volume Rendering
: 3D 영상 데이터의 interactive visualization을 제공한다.

#### Volumes
: 다양한 볼륨 유형들의 외관을 변경하는데 사용된다.
