# About 
___

OpenIGTLink 는 image-guided intervention(이미지-유도 개입)을 위해 특별시 설계된, open-source network communication interface 이다. 

이는 상이한 vendor들을 가지는 imagers(영상장치들), sensors(센서들), surgical robots(수술 로봇들) 그리고 컴퓨터들이 협력적으로 작동하는 image-guided interventions(이미지-유도 개입)을 위해, operating room(ORs, 수술실)에 plug-and-play unified real-time communications(URTC)를 제공하는 것을 목표로 한다.

이 URTC는 이러한 구성요소들 간의 원활한 데이터 흐름을 보장하고, planning(계획), control(제어), delivery(전달), feedback(피드백)을 위한 closed-loop 프로세스르를 가능하게 한다. 

URTC의 적용 사례들은 포함되어있지만 이는 아래의 것들에 국한되지 않는다.

+ 광학 위치 센서와 의료영상 시각화 소프트웨어를 사용한 stereotactic surgical guidance(엄격한 수술 지도)
+ 실시간 MRI와 의료영상 시각화 소프트웨어를 사용한 intraoperative image guidance(수술중 이미지 안내)
+ 로봇 장치 및 수술 계획 소프트웨어를 사용한 robot-assisted interventions(로봇 지원 개입).

> OpenIGTLink의 사양은 오픈되어 있으며, license fee를 내지않고 사용할 수 있다. 
>
> OpenIGTLink는 산업 및 학술 개발자들 모두에게 적합하다. 또한 C/C++ 라이브러리로서 프로토콜의 reference implement(참조 구현)을 제공한다.

# For users
___

: OpenIGTLink는 시술실안의 로컬 영역 네트워크에 연결된 컴퓨터와 장치간에 실시간 데이터 공유를 가능하게 하는 네트워크 통신 인터페이스이다.

이 페이지에는 OpenIGTLink 인터페이스와 함께 제공되는 응용프로그램 소프트웨어 패키지가 나열되어 있다.

Tutorial 페이지에서는 다양한 소프트웨어 패키지에서 OpenIGTLink를 사용하는 방법을 배울수 있는 온라인 tutorial links를 제공한다.

# Research Software Packages
## 3D Slicer

> Web page : http://www.slicer.org

Slicer, 즉 3D Slicer은 시각화 및 이미지 분석을 위한 open-source software이다. 

3D-slicer는 기본적인 windows, linux 및 Mac Os X. 3D Slicer 3.x and 4.x와 함께 외부 소프트웨어 및 장치와 다양한 데이터를 교환할 수 있는 OpenIGTLink 인터페이스 모듈과 함께 제공 된다. 

3D Slicer 4.6에 대한 OpenIGTLink IF 모듈의 설명서를 참조하라. 

(우리가 사용하는 버전은 4.7.0 으로 홀수 번이 들어간 경우 전버전의 development version이다.)

See documentation for OpenIGTLink IF module for 3D Slicer 4.6.(https://www.slicer.org/wiki/Documentation/4.6/Modules/OpenIGTLinkIF)

## Documentation/4.6/Modules/OpenIGTLinkIF
____

### Module Description
___

: OpenIGTLink interface module은 OpenIGTLink 프로토콜을 사용하여, 외부의 소프트웨어나 하드웨어의 네트워크 커뮤니케이션을 위한 program module 이다. 이 모듈은 다음과 같은 특징들을 제공한다.

+ Data import : position, linear transform 및 이미지 데이터를 OpenIGTLink-compliant software 에서 MRML scene으로 가져올 수 있다.
+ Data export : linear transform 과 이미지 데이터를 MRML scene에서 외부 소프트웨어로 내보낼 수 있다.
+ Multi-connection : 동시에 다수의 OpenIGTLink 연결들을 관리할 수 있다. 
+ Locator visualization : 사용자는 MRML scene안의 linear transform 중 하나를 골라, 그것의 position과 orientation을 3D 공산안에 시각화 할 수 있다.
+ slice driving : MRML scene의 linear transform을 기반으로 volume re-slicing 평면을 제어할 수 있다.

![module_description.png](./module_description.png)

#### OpenIGTLink 
: 이는 다양한 image-guided therapy(IGT) 적용을 위해 수술실(OR)에 있는 컴퓨터와 장치간의 통신을 위한 표준화된 매커니즘을 제공한다. 또한 이는 LAN(Local Area Network)의 데이터 교환에 사용되는 디지털 메세징 형식과 프로토콜(규칙들)의 집합이다.

#### MRML
: 이는 slicer에서 사용되는 scene 설명이며, 데이터 모델의 maps이다. MRML 라이브러리는 의료영상 데이터 유형 (volumes, models, transforms, fiducials, cameras, etc)과 그들의 시각화를 관리하기 위한 API를 제공한다.
